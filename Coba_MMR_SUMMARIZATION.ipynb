```python
article = Article('https://www.cnet.com/tech/computing/microsoft-launched-windows-11-now-what-essential-info-about-new-os/', 'https://www.cnbc.com/2021/06/29/windows-11-insider-build-whats-new.html')
article.download()
article.parse()
article.nlp()
text = article.text
print(text)
```

```python
# word_tokenize: memisahkan karakter khusus pada kata (,./@#$%&* dll) dan memisahkan kata pada text


tokens = word_tokenize(text)
print(tokens)
```

```python
## BAg OF WORD
word_frequencies = {}
for word in tokens:
  if word not in word_frequencies:
    word_frequencies[word] =1
  else:
    word_frequencies[word] += 1

print(word_frequencies)
```

```python
#menghapus kata yang termasuk kedalam stopwords
stop_words = list(word_frequencies.keys())
for word in stop_words:
  if word in en_stop_words:
    del word_frequencies[word]

#update kata unik 
stop_words =list(word_frequencies.keys())

len(word_frequencies.keys())
```

```python
#Stemming and lemmatization
stemmer  = SnowballStemmer('english')
lemmatizer = WordNetLemmatizer()

# Fungsi proses stemming kata

def cleanData(sentence):
	sentence = re.sub('[^A-Za-z0-9 ]+', '', text)
	sentence = filter(None, re.split("[.!?", text))
	ret = []
	sentence = stemmer.stem(sentence)	
	for word in sentence.split():
		if not word in stop_words:
			ret.append(word)
	return " ".join(ret)
```

```python
# Fungsi untuk menghitung cosine similarity

def calculateSimilarity(sentence, doc):
	
	if doc == []:
		return 0
	
	vocab = {}
	
	for word in sentence: 
		vocab[word] = 0
	
	docInOneSentence = '';
	for t in doc:
		docInOneSentence += (t + ' ')
		for word in t.split(): 
			vocab[word]=0	
	
	cv = CountVectorizer(vocabulary=vocab.keys())

	docVector = cv.fit_transform([docInOneSentence])
	sentenceVector = cv.fit_transform([sentence])
	return cosine_similarity(docVector, sentenceVector).flatten()
```

```python
# import time
# start = time.time()

%%time

sentences = []
clean = []
originalSentenceOf = {}

#Data cleansing
for line in text:
	parts = line.split('.')
	for part in parts:
		cl = cleanData(part)
		#print cl
		sentences.append(part)
		clean.append(cl)
		originalSentenceOf[cl] = part		

setClean = set(clean)
```

```python
#menghitung cosine similiarity	
scores = {}
for data in clean:
	temp_doc = setClean - set([data]) 
	score = calculateSimilarity(data, list(temp_doc))
	scores[data] = score
	#print score
	print(scores)
```

```python
# calculate MMR
n = 20 * len(sentences) / 100 #Set mau seberapa banyak hasil dari summarisasi
alpha = 0.5 #Set Nilai Alpha
summarySet = [] #Set list kosong untuk menyimpan hasil teks summarisasi

while n > 0: #melakukan looping selama nilai n (banyak kalimat yg akan diekstraksi) lebih besar dari 0
	mmr = {} #Inisialisasi dictionary untuk menyimpan nilai MMR 
	#kurangkan dengan set summary
  # Removed dict.iteritems(), dict.iterkeys(), and dict.itervalues().
  # Instead: use dict.items(), dict.keys(), and dict.values() respectively.

	for sentence in scores.keys():
		if not sentence in summarySet:
			mmr[sentence] = alpha * scores[sentence] - (1-alpha) * calculateSimilarity(sentence, summarySet) #Mencari nilai MMR tiap kalimat yg ada pada document terhadap text yg sudah dirangkum
	 
	selected = max(mmr.items(), key=operator.itemgetter(1))[0]	#Memilih teks dengan nilai MMR paling besar dari proses sebelumnya
	
	summarySet.append(selected) #Menyimpan text tersebut kedalam list yg menyimpan hasil text sunmarisasi
	print(summarySet) #Melihat text apa saja yg sudah masuk kedalam hasil summarisasi
	n -= 1

#rint str(time.time() - start)
```
